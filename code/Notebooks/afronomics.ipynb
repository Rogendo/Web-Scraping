```python
from bs4 import BeautifulSoup
import numpy as np
from selenium import webdriver
import time
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

               
```

```python
# load the webdriver
options = webdriver.ChromeOptions()
options.add_argument('headless')
options.add_argument('disable-gpu')

driver = webdriver.Chrome(options=options)

```

```python
topics = ['academic-forum', 'afsdjn', 'journal', 'analysis', 'news-and-events', 'repository', 'video-content']
BASE_URL = 'https://www.afronomicslaw.org/index.php'

def scrape_afronomicslaw(topic, page=1):
    url = f'https://www.afronomicslaw.org/index.php/category/{topic}?page={page}'
    data = []
    try:
        driver.get(url) 
        time.sleep(5)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        articles = soup.find_all('article', class_='node node--type-article node--promoted node--view-mode-teaser clearfix')
        
        if not articles:
            print(f'No articles found for {topic} on page {page}')
            return pd.DataFrame()  # Return empty dataframe if no articles found
        
        for article in articles:
            title = article.find('span', class_="field field--name-title field--type-string field--label-hidden")
            title = title.text if title else 'NaN'
            summary = article.find('div', class_='clearfix text-formatted field field--name-body field--type-text-with-summary field--label-hidden field__item')
            summary = summary.text if summary else 'NaN'
            link = article.find('a')['href']
            link = BASE_URL + link if link else 'NaN'
            tags = article.find('div', class_='field field--name-field-tags field--type-entity-reference field--label-above clearfix')
            tags = tags.text if tags else 'NaN'
            
            driver.get(link)
            time.sleep(5)
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            content = soup.find('div', class_='clearfix text-formatted field field--name-body field--type-text-with-summary field--label-hidden field__item')
            content = content.text if content else 'NaN'
            
            data.append({
                'title': title,
                'tags': tags,
                'summary': summary,
                'content': content,
                'link': link
            })
        
    

    except Exception as e:
        print(f'Error: {e}')
    df = pd.DataFrame(data)
    return df

import os

for topic in topics:
    for i in range(1, 10):
        df = scrape_afronomicslaw(topic, i)
        if not df.empty:
            if os.path.exists('afronomicslaw.csv') and os.path.getsize('afronomicslaw.csv') > 0:
                df.to_csv('afronomicslaw.csv', mode='a', header=False, index=False)
            else:
                df.to_csv('afronomicslaw.csv', mode='w', header=True, index=False)
        time.sleep(5)
        print('Done')

```
